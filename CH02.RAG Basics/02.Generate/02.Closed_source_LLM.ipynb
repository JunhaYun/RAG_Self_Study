### OpenAI API 활용

```python
# OpenAI API 사용 예제
import openai
from IPython.display import display, Markdown # 답변을 마크다운 형식으로 보여준다
# API 키 설정
client = openai.OpenAI()

# GPT-4o 모델 호출
# client.chat.completions => 그냥 챗봇 기능
response_gpt4o = client.chat.completions.create(
    model="gpt-4o",
    temperature=0.7,
    messages=[
        {"role": "system", "content": "당신은 반도체 회사 MES 팀 엔지니어입니다."}, # 생략 가능
        {"role": "user", "content": """"
        MES 시스템에서 발생하는 오류를 엔지니어가 직접 로그를 하나씩 디버깅하지 않고,
        AI Assistant가 자동으로 원인을 탐색하고 대응 방향을 제안하는 시스템을 설계해줘.

        조건:
        - MES 이벤트 로그와 설비 로그를 활용한다.
        - 과거 장애 대응 이력을 검색한다.
        - 오류 메시지와 유사 사례를 기반으로 원인 후보를 제시한다.
        - 관련 매뉴얼/PDF 문서를 RAG로 검색한다.
        - 최종적으로 엔지니어용 대응 리포트를 생성한다.

        다음 항목을 포함해줘:
        1. 시스템 아키텍처
        2. 데이터 흐름
        3. RAG 구성 방식
        4. LLM/Agent의 역할
        5. 실제 제조 현장 적용 시 한계와 주의사항
        """}
    ]
)

display(Markdown(f"###답변\n\n{response_gpt4o.choices[0].message.content}"))
```

```python
# o3-mini 모델 호출 (추론 모델)
# client.responses => agent API 기능
response_o3mini = client.responses.create(
    model="o3-mini",  # o3-mini-high가 아닌 o3-mini 사용
    reasoning={"effort": "low"}, #low, medium, high  => 추론을 얼마나 강하게 할 것인가?
    input=[
        {"role": "system", "content": "당신은 유용한 AI 비서입니다."},
        {"role": "user", "content": """
        
        MES 시스템에서 발생하는 오류를 엔지니어가 직접 로그를 하나씩 디버깅하지 않고,
        AI Assistant가 자동으로 원인을 탐색하고 대응 방향을 제안하는 시스템을 설계해줘.

        조건:
        - MES 이벤트 로그와 설비 로그를 활용한다.
        - 과거 장애 대응 이력을 검색한다.
        - 오류 메시지와 유사 사례를 기반으로 원인 후보를 제시한다.
        - 관련 매뉴얼/PDF 문서를 RAG로 검색한다.
        - 최종적으로 엔지니어용 대응 리포트를 생성한다.

        다음 항목을 포함해줘:
        1. 시스템 아키텍처
        2. 데이터 흐름
        3. RAG 구성 방식
        4. LLM/Agent의 역할
        5. 실제 제조 현장 적용 시 한계와 주의사항
        """}
    ]
)
```

### Claude API 활용하기

```python
# Claude API 사용 예제
import anthropic

# API 키 설정
client = anthropic.Anthropic()

from IPython.display import display, Markdown

# 클로드의 경우 block 단위
def render_response(blocks):
    for block in blocks:
        if getattr(block, "type", None) == "thinking":
            display(Markdown(f"### 🤔 Thinking\n\n{block.thinking}")) # 추론 block
        elif getattr(block, "type", None) == "text":
            display(Markdown(f"### 💬 답변\n\n{block.text}")) # 답변 block

# Claude 3.7 Sonnet 호출 예시
message = client.messages.create(
    model="claude-3-7-sonnet-20250219",  # Claude 3.7 Sonnet 모델
    max_tokens=1000,
    temperature=0.7,
    system="당신은 유용한 AI 비서입니다.",
    messages=[
        {"role": "user", "content": "파이썬으로 피보나치 수열을 계산하는 코드를 작성해주세요."}
    ]
)

# 응답 출력
render_response(message.content)
```

```python
response = client.messages.create(
    model="claude-3-7-sonnet-20250219",
    max_tokens=3000,
    thinking={
        "type": "enabled",
        "budget_tokens": 2000 # 길이가 길어질수록 추론을 잘한다(?)
    },
    messages=[
        {"role": "user", "content": "파이썬으로 피보나치 수열을 계산하는 코드를 작성해주세요."}
    ]
)

render_response(response.content)
```

### Gemini API 활용하기

```python
# Gemini API 사용 예제
from google import genai

# API 키 설정
client = genai.Client()

# Gemini 모델 호출 (올바른 모델명 사용)
response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents=["파이썬으로 이진 검색 알고리즘을 구현해주세요."],
)

print("Gemini 2.0 Flash 응답:")
display(Markdown(f"### 💬 답변\n\n{response.text}"))
```

### Langchain을 활용한 Closed LLM 활용법

#### OpenAI o3-mini

```python
import os
from langchain_openai import ChatOpenAI

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
o3_mini = ChatOpenAI(model="o3-mini")

for token in o3_mini.stream(
    "내가 가장 좋아하는 동물은 물범이야. 물범에 대해 간략한 정보를 줘.",
    reasoning_effort="medium"
    ):
    print(token.content, end="")
```
