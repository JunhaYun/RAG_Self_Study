### PDF 파일에서 문서 로드하기
- LangChain의 PyPDFLoader를 이용

```python
file_path = "data/PIKC.pdf" # 예시논문은 내가 썼던 논문으로..

# LangChain의 PyPDFLoader를 사용하여 PDF 파일 읽기
from langchain.document_loaders import PyPDFLoader

# pypdf로 PDF 객체 생성
loader = PyPDFLoader(
    file_path,
    # "plain" : 그냥 텍스트만 순서대로 추출
    # "layout" : PDF의 레이아웃을 고려하여 텍스트 추출
    extraction_mode="plain",   
    #extraction_mode="layout",                  
    )

# 비동기 방식으로 pdf의 각 페이지를 로드하며 pages 리스트에 추가
# 비동기 방식 : PDF 읽는 동안 API 호출, 웹 요청 등 다양한 작업 동시에 가능

# 읽는 방식에는 크게 2가지가 있다
# 1. aload : PDF 전체를 한 번에 메모리로 로드 - 코드 단순하지만 큰 pdf는 메모리 부담이 될 수 있음
docs = await loader.aload()
# 2. alazy_load : 한 장씩 넘겨주면서 전달 - 메모리 부담 줄이지만 코드 복잡해질 수 있음
#pages = []
#async for page in loader.alazy_load():
#    pages.append(page)

print(docs[0].page_content[:100]) # PDF의 첫 페이지에서 처음 100자만 출력
print(docs[0].metadata) # 생성날짜, 저자, 키워드 등 PDF 부가정보 (metadata) 출력

#print(f"{pages[0].metadata}\n")
#print(pages[0].page_content[:500]) 
```
